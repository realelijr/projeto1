```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import seaborn
```

```python
teste = pd.read_csv('2019.csv', sep=';')
```

```python
teste.columns = ['regiao', 'uf', 'cidade', 'sexo', 'faixaEtaria', 'dataFinalizacao', 'tempoResposta', 'nomeFantasia', 'segmentoMercado', 'area', 'assunto', 'grupoProblema', 'problema', 'formaContratacao', 'procurouEmpresa', 'respondida', 'situacao', 'avaliacaoReclamacao', 'notaConsumidor', 'total']
```

```python
teste.head(1)
```

```python
teste.count()
```

```python
teste['notaConsumidor'].mode()
```

```python
teste_original = teste.copy() 
```

```python
teste['notaConsumidor'].fillna(teste['notaConsumidor'].mode()[0], inplace=True)
```

```python
teste.count()
```

```python
teste['tempoResposta'].fillna(teste['tempoResposta'].mode()[0], inplace=True)
```

```python
teste = teste.drop(['nomeFantasia', 'dataFinalizacao', 'cidade', 'problema', 'segmentoMercado','assunto','regiao'], axis = 1)
```

```python
teste.head()
```

```python
teste.dtypes
```

```python
teste['sexo'] = teste['sexo'].apply(lambda x: 1 if x == 'male' else 0)
```

```python
teste = pd.get_dummies(teste)
```

```python
teste.head()
```

```python
alvo = teste["notaConsumidor"] 
```

```python
teste =  teste.drop('notaConsumidor', axis = 1)
teste.columns
```

```python
from sklearn.model_selection import train_test_split

X_treino, X_teste, Y_treino, Y_teste = train_test_split(teste, alvo, test_size = 0.22, random_state = 0)
```

```python
X_treino.describe()
```

```python
Y_treino.describe()
```

```python
resultados ={}
```

```python
# Gaussian Naive Bayes
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

gaussian = GaussianNB()
gaussian.fit(X_treino, Y_treino)
Y_pred = gaussian.predict(X_teste)
acc_gaussian = round(accuracy_score(Y_pred, Y_teste) * 100, 2)
resultados[acc_gaussian] = gaussian
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

logreg = LogisticRegression()
logreg.fit(X_treino, Y_treino)
Y_pred = logreg.predict(X_teste)
acc_logreg = round(accuracy_score(Y_pred, Y_teste) * 100, 2)
resultados[acc_logreg] = logreg
```

```python
# Support Vector Machines
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

svc = SVC()
svc.fit(X_treino, Y_treino)
Y_pred = svc.predict(X_teste)
acc_svc = round(accuracy_score(Y_pred, Y_teste) * 100, 2)
resultados[acc_svc] = svc
```

```python
# Linear SVC
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

linear_svc = LinearSVC()
linear_svc.fit(X_treino, Y_treino)
Y_pred = linear_svc.predict(X_teste)
acc_linear_svc = round(accuracy_score(Y_pred, Y_teste) * 100, 2)
resultados[acc_linear_svc] = linear_svc
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

decisiontree = DecisionTreeClassifier()
decisiontree.fit(X_treino, Y_treino)
Y_pred = decisiontree.predict(X_teste)
acc_decisiontree = round(accuracy_score(Y_pred, Y_teste) * 100, 2)
resultados[acc_decisiontree] = decisiontree
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

randomforest = RandomForestClassifier()
randomforest.fit(X_treino, Y_treino)
Y_pred = randomforest.predict(X_teste)
acc_randomforest = round(accuracy_score(Y_pred, Y_teste) * 100, 2)
resultados[acc_randomforest] = randomforest
```
